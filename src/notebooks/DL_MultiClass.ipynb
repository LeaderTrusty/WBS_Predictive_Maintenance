# Baseline model

```python
early_stopping = EarlyStopping(
    min_delta=0.001, # minimium amount of change to count as an improvement
    patience=40, # how many epochs to wait before stopping
    restore_best_weights=True,
)
```

```python
input_shape = X.shape[1]
n_labels = y.shape[1]
```

```python
def baseline_model(input_shape, n_nodes=128, n_hidden=2, n_labels=5):
    ''' create model '''

    model = Sequential()
    model.add(Dense(n_nodes, activation='relu', input_shape=[input_shape]))
    for i in range(n_hidden):
        model.add(Dropout(0.3))
        model.add(BatchNormalization())
        model.add(Dense(n_nodes, activation='relu'))
    model.add(Dropout(0.3))
    model.add(BatchNormalization())
    model.add(Dense(n_labels, activation='sigmoid'))

    return model
```

```python
def run_model(X_train, y_train, X_test, y_test, n_nodes, n_hidden, input_shape, n_epochs, n_labels):
    ''' compute deep learning model(s)'''

    model = baseline_model(input_shape)

    model.compile(
        optimizer='adam',
        loss='binary_crossentropy',
        metrics=['binary_accuracy']
    )

    history = model.fit(
        X_train, y_train,
        validation_data=(X_test, y_test),
        batch_size=int(n_nodes/2),
        epochs=n_epochs,
        callbacks=[early_stopping],
        verbose=0
    )

    predictions = model.predict(X_test)
    predictions = pd.DataFrame(predictions, index=y_test.index, columns=y_test.columns)

    return history, model, predictions
```

```python
history, model, predictions = run_model(X_train, y_train, X_test, y_test, 128, 2, X_train.shape[1], 200, 5)

history_df = pd.DataFrame(history.history)
```

```python
pred_vals = pd.DataFrame(predictions.round(), index=y_test.index, columns=y_test.columns)
pred_vals.describe()
```

```python
accuracy = accuracy_score(y_test, pred_vals)
print(accuracy)
```

```python
for i in range(y_test.shape[1]):
    accuracy = accuracy_score(y_test.iloc[:, i], pred_vals.iloc[:, i])
    print(f'{y_test.columns[i]} - {100*accuracy} %')
```
